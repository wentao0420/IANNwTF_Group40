```python
class Layer(object):
    '''Layers of a perceptron'''
    
    def __init__(self, n_units, input_units):
        self.bias_vector = np.empty(n_units)
        self.weight_matrix = np.random.rand([input_units, n_units])
        self.w_b = self.weight_matrix.append(self.bias_vector)
        self.layer_input = None
        self.layer_pre = None
        self.layer_act = None
        self.layer_targets = None
```

2. A method called 'forward_step', which returns each unit’s activation
(i.e. output) using ReLu as the activation function.

```python
# relu (activation function)
def relu(x):
    for i in range(len(x)):
        if x[i] > 0:
            return x[i]
        else:
            return 0

# relu derivative
def relu_d(x):
    for i in range(len(x)):
        if x[i] > 0:
            return 1
        else:
            return 0
```

3. A method called 'backward_step', which updates each unit’s parameters
(i.e. weights and bias).

```python
def forward_step(self):
    self.layer_act = relu(self.layer_pre @ self.w_b)
        
def backward_step(self, layer_targets, l_rate=0.03):
    self.layer_targets = layer_targets
    input_transposed = np.transpose(self.layer_input)
    weights_transposed = np.transpose(self.weight_matrix)
        
    dL_dactivation = self.layer_act - self.layer_targets
        
    # compute the gradients
    dL_dWl = input_transposed @ ( np.multiply( relu_d(self.pre_layer), dL_dactivation ))
    dL_dbl = np.multiply( relu_d(self.pre_layer), dL_dactivation )
    dL_dinput = (relu_d(self.pre_layer) * dL_dactivation) @ weights_transposed
        
    # update the layer's parameters
    self.weight_matrix = self.weight_matrix - l_rate*dL_dWl
    self.bias_vector = self.bias_vector - l_rate*dL_dbl
```
